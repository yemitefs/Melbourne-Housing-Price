# Starting the Preprocessing
The first thing I have to do is to extract the month from the date the property was sold as it plays a key factor in determining the price of houses.

```python
# Get Months out of Dates
data['Date'] = pd.to_datetime(data['Date'])
column = data['Date']
feature_col = pd.DataFrame({"month":column.dt.month})
data =  pd.concat([data,feature_col], axis = 1)
```

```python
data.head()
```

```python
data=data.drop(['Address','Postcode','Date','SellerG','Suburb','BuildingArea','YearBuilt'], axis=1)
```

```python
# to confirm if columns have been dropped
data.head()
```

```python
#statistical summary of our numerical columns
data.describe()
```

```python
#to check for null-data
print(data.isnull().sum())
```

```python
data.loc[data.Car.isnull(), 'Car']  = data.groupby('Type')['Car'].transform(lambda x: x.fillna(x.mean()))
```

# Data Modelling

```python
y=data.Price
```

```python
X=data.drop(['Price'], axis=1)
```

```python
cols=['Type','Method','CouncilArea','Regionname']
```

```python
for item in cols:
    X[item] = X[item].astype("category").cat.codes +1
```

```python
X[cols]=X[cols].astype(np.int64)
```

```python
X.columns
```

```python
X.dtypes
```

```python
print (X['Regionname'].unique())
```

```python
X.head()
```

```python
cat_features_index=[1,2,8,11]
```

```python
clf=CatBoostRegressor()
```

```python
clf.fit(X,y, cat_features=cat_features_index)
```

```python
best_result = clf.best_score_  
print(best_result)
```

```python
clf.score(X,y)
```

```python
scores = -1 * cross_val_score(clf, X, y,
                              cv=3,
                              scoring='neg_mean_absolute_error')

print("MAE scores:\n", scores)
```

```python
from sklearn.externals import joblib

joblib.dump(clf, 'model.pkl')
```
